# Stage 3 evaluation skeleton

Evaluate prediction JSONL files against Stage 2 nvBench post-query examples and write metrics under the canonical Drive root.

```python
from __future__ import annotations

import os
import subprocess
import sys
from pathlib import Path

DRIVE_ROOT = Path(os.environ.get('T2V_DRIVE_ROOT', '/content/drive/MyDrive/diploma/petr_text_to_visualization_part'))
REPO_ROOT = Path(os.environ.get('T2V_REPO_ROOT', '/content/petukhov_t2v_repo'))

os.environ['T2V_DRIVE_ROOT'] = str(DRIVE_ROOT)
os.environ['T2V_REPO_ROOT'] = str(REPO_ROOT)

if not (REPO_ROOT / 'pyproject.toml').exists():
    raise RuntimeError(f'Repository checkout not found: {REPO_ROOT}')

subprocess.check_call([sys.executable, '-m', 'pip', 'install', '-q', '-e', str(REPO_ROOT)])
print({'drive_root': str(DRIVE_ROOT), 'repo_root': str(REPO_ROOT)})
```

```python
from __future__ import annotations

import os
import subprocess
import sys
from pathlib import Path

drive_root = Path(os.environ['T2V_DRIVE_ROOT'])
repo_root = Path(os.environ['T2V_REPO_ROOT'])
examples = drive_root / 'datasets' / 'processed' / 'nvbench_postquery' / 'examples_sample200.jsonl'
predictions = drive_root / 'runs' / 'stage3_placeholder' / 'predictions.jsonl'
output_dir = drive_root / 'runs' / 'stage3_placeholder' / 'metrics'

if not predictions.exists():
    raise RuntimeError(f'Predictions JSONL not found yet: {predictions}')

cmd = [
    sys.executable,
    str(repo_root / 'scripts' / 'evaluate_predictions.py'),
    '--examples', str(examples),
    '--predictions', str(predictions),
    '--output-dir', str(output_dir),
    '--json',
]
print('RUN', ' '.join(cmd))
result = subprocess.run(cmd, text=True, capture_output=True)
print(result.stdout)
if result.stderr:
    print(result.stderr)
if result.returncode != 0:
    raise RuntimeError(f'evaluation failed with exit code {result.returncode}')
print('STAGE3_EVALUATION_OK')
```
